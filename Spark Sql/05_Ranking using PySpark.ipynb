```
order_df=spark.read.csv("dbfs:/public/retail_db1/orders",schema='order_id INT,order_date DATE,order_customer_id INT,order_status STRING')
```

```
order_items_df=spark.read.csv('dbfs:/public/retail_db1/order_items',
    schema='order_item_id INT,order_item_order_id INT,order_item_product_id INT,order_item_quantity INT,order_item_subtotal FLOAT,order_item_product_price FLOAT')
```

```
from pyspark.sql.functions import round,sum,col
```

```
daily_revenue=order_df.\
    filter("order_status IN ('COMPLETE','CLOSED')").\
        join(order_items_df,order_df['order_id']==order_items_df['order_item_order_id']).\
            groupBy(order_df['order_date']).\
                agg(round(sum('order_item_subtotal'),2).alias('revenue'))
            
daily_product_revenue=order_df.\
    filter("order_status IN ('COMPLETE','CLOSED')").\
        join(order_items_df,order_df['order_id']==order_items_df['order_item_order_id']).\
            groupBy(order_df['order_date'],order_items_df['order_item_product_id']).\
                agg(round(sum('order_item_subtotal'),2).alias('revenue'))
```

```
display(daily_revenue.orderBy('order_date'))
```

```
display(daily_product_revenue.orderBy('order_date',col('revenue').desc()))
```

```
from pyspark.sql.functions import dense_rank,col,rank,date_format
from pyspark.sql.window import Window
```

```
display(
    daily_product_revenue. \
    filter("order_date='2014-01-01'").\
        withColumn('drnk',dense_rank().over(Window.orderBy(col('revenue').desc()))).\
            orderBy('order_date',col('revenue').desc())
)

    
```

```
#only getting Top 5 rnk
display(
    daily_product_revenue. \
    filter("order_date='2014-01-01'").\
        withColumn('drnk',dense_rank().over(Window.orderBy(col('revenue').desc()))).\
            filter("drnk<=5").\
            orderBy('order_date',col('revenue').desc())
)
```

```
display(
    daily_product_revenue.\
        filter("date_format(order_date,'yyyyMM')=201401").\
            groupBy(date_format('order_date','yyyyMM').alias('order_month'),'order_item_product_id').\
                agg(round(sum('revenue'),2).alias('revenue')).\
                    withColumn('drnk',dense_rank().over(Window.orderBy(col('revenue').desc()))).\
                        filter("drnk<=5")
)
```

```
display(
    daily_product_revenue.\
        filter("date_format(order_date,'yyyyMM')=201401").\
            withColumn('drnk',dense_rank().over(Window.partitionBy('order_date').orderBy(col('revenue').desc()))).\
                orderBy('order_date',col('revenue').desc())
)
```
